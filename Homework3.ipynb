```python

#Importing pandas and numpty
import pandas as pd
import numpy as np
```

```python
#Reading csv. file gave issue so i had to deleter everything but date
yahoo = pd.read_csv('/yahoo.csv')

#displaying data
yahoo
```

```python
#Showing dates of data
dates=yahoo['date']
dates
```

```python
#Converting dataset for calculations
yahoo.dtypes
data_date = pd.to_datetime(yahoo['date'], format='%m/%d/%Y')

data_date
```

```python
#Showing the max date in dataset
data_date.max()
```

```python
#Adding new column to dataset for difference in days
yahoo["Date_Difference(Days)"] = (data_date - data_date.max())

yahoo
```

```python
#converting days to months in new column
yahoo["Date_Difference(Months)"] = yahoo["Date_Difference(Days)"] / np.timedelta64(1, 'M')

yahoo
```

```python
#New Dataset saved as a csv

yahoo.to_csv(r'yahoo_clean.csv', index=False)
```
